**Создаём функции**

```python
def duplicates(name, df, data_output):
    "Выводит количество дубликатов строк, для вывода дубликатов нужно ввести data_output = 1"
    print()
    if df.duplicated().sum() > 0:
        print(f'Количество дубликатов в таблице "{name}" равно {df.duplicated().sum()} шт.')
        if data_output == 1:
            display(df[df.duplicated()])
    else:
        print(f'Дубликаты в таблице "{name}" отсутствуют')
    print('---------------------------------------------------------------')
```

```python
def info(name, df):
    "Выводит информацию о таблице"
    print()
    print(f'\033[1mВыводим информацию о таблице "{name}":\033[0m')
    print('---------------------------------------------------------------')
    print(f'- Выводим первые 5 строк таблицы')
    display(df.head())
    print()
    print(f'- Выводим общую информацию таблицы')
    print()
    df.info()
    print()
    print(f'- Выводим характеристики данных таблицы')
    display(df.describe().T)
    print()
    print(f'- Таблица "{name}" содержит {df.shape[0]} строк и {df.shape[1]} столбцов')
    print('---------------------------------------------------------------')
```

```python
def make_features(df, feature, max_lag, rolling_size):
    "Создаёт дополнительные признаки"
    for i in range(max_lag):
        df['lag_'+feature+'_'+str(i+1)] = df[feature].shift(i+1)
    for j in [1, 2, 4]:
        df['rolling_mean_'+feature+'_'+str(rolling_size*j)] = df[feature].rolling(rolling_size*j).mean()
        df['rolling_std_'+feature+'_'+str(rolling_size*j)] = df[feature].rolling(rolling_size*j).std()
```

```python
def missing_values(name, df, data_output):
    "Вывод информации о пропущенных данных, для вывода строк с пропущенными данными нужно ввести data_output = 1"
    print()
    if sum(df.isna().sum()) > 0:
        print(f'Пропущенные данные таблицы "{name}":')
        print()
        for column in df.columns:
            if df[column].isna().sum() > 0:
                print()
                print(f'- количество пропущенных данных столбца ["{column}"] = {df[column].isna().sum()} шт.')
                if data_output == 1:
                    display(df[df[column].isna()])
    else:
        print(f'В таблице "{name}" отсутствуют пропущенные данные')
    print('---------------------------------------------------------------')
```

**Добавляем дополнительные признаки в таблицу**

```python
data['week_day'] = data.index.weekday # день недели
data['decade'] = data.index.to_series().apply(lambda x: 1 if x.day < 11 else 2 if 10 < x.day < 21 else 3) # декада месяца
data['month'] = data.index.month # месяц
data['quarter'] = data.index.to_series().apply(lambda x: 1 if x.month < 4 else 2 if 3 < x.month < 7 else 3 if 6 < x.month < 10 else 4) # квартал
data['half_year'] = data.index.to_series().apply(lambda x: 1 if x.month < 7 else 2) # полугодие

data['close_open'] = data['close'] - data['open'] # разность цен между закрытием и открытием
data['high_low'] = data['high'] - data['low'] # разность цен между high и low
data['profit'] = data['close_open'].abs() * data['vol'] # прибыль
data = data.query('profit > 0')
data['exceeding'] = data['high_low'] - data['close_open'].abs() # разность цен между максимальным колебанием и дневным
data = data.query('exceeding > 0')

data['close_open_cat'] = data.apply(lambda row: 0 if row['close'] < row['open'] else 1 if row['open'] < row['close'] else 3, axis=1) # категория: рост или снижение
data['exceeding_cat'] = data.apply(lambda row: 0 if abs(row['close_open']) < row['exceeding'] else 1 if row['exceeding'] < abs(row['close_open']) else 3, axis=1) # категория: рост или снижение

data['close_change'] = data['close'].pct_change() # изменение цены закрытия
data['vol_change'] = data['vol'].pct_change() # изменение объёма
data['profit_change'] = data['profit'].pct_change() # изменение прибыли
data['exceeding_change'] = data['exceeding'].pct_change() # изменение соотношения цен

data['close_change_cat'] = data.apply(lambda row: 0 if row['close_change'] < 0 else 1 if row['close_change'] > 0 else 3, axis=1) # категория: рост или снижение
data['vol_change_cat'] = data.apply(lambda row: 0 if row['vol_change'] < 0 else 1 if row['vol_change'] > 0 else 3, axis=1) # категория: рост или снижение
data['profit_change_cat'] = data.apply(lambda row: 0 if row['profit_change'] < 0 else 1 if row['profit_change'] > 0 else 3, axis=1) # категория: рост или снижение
data['exceeding_change_cat'] = data.apply(lambda row: 0 if row['exceeding_change'] < 0 else 1 if row['exceeding_change'] > 0 else 3, axis=1) # категория: рост или снижение
```

**Добавляем дополнительные признаки, идаляем признаки, чтобы избежать мультиколлинеарность**

```python
for a in ['open', 'close', 'vol']: # отстающие значения и скользящие средние
    make_features(data, a, 10, 10)

# технические индикаторы:
data['rsi'] = ta.momentum.RSIIndicator(data['close']).rsi()
data['macd'] = ta.trend.macd_diff(data['close'])
bb = ta.volatility.BollingerBands(data['close'])
data['bb_upper'] = bb.bollinger_hband()
data['bb_lower'] = bb.bollinger_lband()
data['bb_position'] = (data['close'] - data['bb_lower']) / (data['bb_upper'] - data['bb_lower'])

data = data.drop(['open', 'high', 'low', 'quarter'], axis=1)
size('data', data)
```

**Выделяем выборки с входными и целевыми тренировочными и тестовыми данными**

```python
train, test = train_test_split(data, shuffle=False, test_size=TEST_SIZE)
train = train.dropna()

X_train = train.drop('target', axis=1)
y_train = train['target']
X_test = test.drop('target', axis=1)
y_test = test['target']

size('X_train', X_train)
size('X_test', X_test)
```

**Создаём списки с названиями признаков и пайплайны для подготовки признаков**

```python
cat_columns = ['week_day', 'decade', 'month', 'half_year', 'exceeding_cat', 'close_open_cat', 'close_change_cat', 'vol_change_cat', 'profit_change_cat', 'exceeding_change_cat']
num_columns = X_train.drop(['week_day', 'decade', 'month', 'half_year', 'exceeding_cat', 'close_open_cat', 'close_change_cat', 'vol_change_cat', 'profit_change_cat', 'exceeding_change_cat'], axis=1).columns.tolist()
num_columns = X_train.columns.tolist()
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))])

data_preprocessor = ColumnTransformer(
    [('ord', ord_pipe, cat_columns),
     ('num', StandardScaler(), num_columns)
    ], remainder='passthrough')
```

**Создаём итоговый пайплайн для подготовки данных и моделей. В качестве моделей будем использовать LGBMRegressor и CatBoostRegressor. Производим обучение моделей, определяем параметры лучшей модели, выводим метрику ROC-AUC на тренировочной выборке**

```python
pipe_final = Pipeline(
    [('preprocessor', data_preprocessor),
     ('models', LGBMClassifier(random_state=RANDOM_STATE))])

param_grid = [
    {'models': [LGBMClassifier(random_state=RANDOM_STATE, verbose=0)],
     'models__num_iterations': [100, 200],
     'models__learning_rate': [0.1, 0.2],
     'preprocessor__num': [StandardScaler(), 'passthrough']},

    {'models': [CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)],
     'models__iterations': [100, 200],
     'models__learning_rate': [0.1, 0.2],
     'preprocessor__num': [StandardScaler(), 'passthrough']}]

tscv = TimeSeriesSplit(n_splits=5)
randomized_search_1 = RandomizedSearchCV(pipe_final, param_grid, cv=tscv, scoring='roc_auc', random_state=RANDOM_STATE, n_jobs=-1)
randomized_search_1.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_1.best_estimator_)
print('Метрика лучшей модели ROC-AUC на тренировочной выборке:', randomized_search_1.best_score_)
```

```python
y_test_pred_1 = randomized_search_1.predict(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_pred_1)}')
```

**Строим матрицу ошибок**

```python
cm = confusion_matrix(y_test, y_test_pred_1)
plt.figure(figsize=(8, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.title('Тепловая карта матрицы ошибок', pad=30)
plt.xlabel('Predicted')
plt.ylabel('True label')
plt.show()
```
